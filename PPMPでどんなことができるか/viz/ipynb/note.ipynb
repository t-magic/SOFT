# data_n_htext の共起(第2キーワード)

```python
xlsx_file = '../foundation_0_htext_fa3_d3w20n100e200f1.xlsx'
```

```python
import pandas as pd
# read xlsx file
df = pd.read_excel(xlsx_file)
df.head()
```

## 不要な列を削除する

```python
df.drop("番号", axis=1, inplace=True)
df.drop("コメント(FA)", axis=1, inplace=True)
df.head()
```

```python
df.columns
```

## 語の共起回数の行列を求める

```python
prod_df = df.T.dot(df)
prod_df
```

```python
prod_df.columns[:5]
```

```python
prod_df.columns[0]
```

```python
sorted_prod_df = prod_df.sort_values(by=prod_df.columns[0], ascending=False, inplace = False) 
sorted_prod_df
```

```python
sorted_prod_df.loc[prod_df.index[:5], prod_df.index[:5]]
```

## 取り上げた語との共起数を求め、共起数の多い順に語をソートする

```python
sorted_prod_df.columns[0]
```

```python
sorted_prod_df[prod_df.columns[0]]
```

```python
vlist = sorted_prod_df[prod_df.columns[0]]
vlist
```

## 第2、第3キーワード

```python
for i in range(len(prod_df.columns)):
    headword = prod_df.columns[i]
    #print(headword)
    sorted_prod_df = prod_df.sort_values(by=headword, ascending=False, inplace = False)
    #print(sorted_prod_df)
    wlist = sorted_prod_df[headword].index.tolist()
    vlist = sorted_prod_df[headword].tolist()
    #print('wlist = {}'.format(wlist))
    #print('vlist = {}'.format(vlist))
    output_list = []
    for count in range(len(wlist)):
        if count > 4:
            break
        #print('{0}({1})'.format(wlist[count].split('(')[0], vlist[count]))
        word = '{0}({1})'.format(wlist[count].split('(')[0], vlist[count])
        if word != headword:
            output_list.append(word)
    output_str = '{}: '.format(headword) + ', '.join(output_list)
    print(output_str)
```
